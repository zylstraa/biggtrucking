### onboard data dictionary
* **Id**: similar to index and not necessary for our data
* **Name**: Variable being measured at the time of whatever fault number FaultId is
* **Value**: the actual measurement value for whatever variable is being measured in Name
* **FaultId**: The same as the fault data csv's RecordID

```python
# let's see if we can fill NaNs with mean by first converting to a float
#there are a lot of commas instead of . making the value act as a string rather than a float 
onboard_df['BarometricPressure'] = onboard_df['BarometricPressure'].str.replace(',','.').astype(float)
onboard_df['AcceleratorPedal'] = onboard_df['AcceleratorPedal'].str.replace(',','.').astype(float)
onboard_df['DistanceLtd'] = onboard_df['DistanceLtd'].str.replace(',','.').astype(float)
onboard_df['EngineCoolantTemperature'] = onboard_df['EngineCoolantTemperature'].str.replace(',','.').astype(float)
onboard_df['EngineOilPressure'] = onboard_df['EngineOilPressure'].str.replace(',','.').astype(float)
onboard_df['EngineOilTemperature'] = onboard_df['EngineOilTemperature'].str.replace(',','.').astype(float)
onboard_df['EngineRpm'] = onboard_df['EngineRpm'].str.replace(',','.').astype(float)
onboard_df['EngineTimeLtd'] = onboard_df['EngineTimeLtd'].str.replace(',','.').astype(float)
onboard_df['FuelLtd'] = onboard_df['FuelLtd'].str.replace(',','.').astype(float)
onboard_df['IntakeManifoldTemperature'] = onboard_df['IntakeManifoldTemperature'].str.replace(',','.').astype(float)
onboard_df['Speed'] = onboard_df['Speed'].str.replace(',','.').astype(float)
onboard_df['SwitchedBatteryVoltage'] = onboard_df['SwitchedBatteryVoltage'].str.replace(',','.').astype(float)
onboard_df['TurboBoostPressure'] = onboard_df['TurboBoostPressure'].str.replace(',','.').astype(float)
onboard_df['FuelLevel'] = onboard_df['FuelLevel'].str.replace(',','.').astype(float)
onboard_df['FuelRate'] = onboard_df['FuelRate'].str.replace(',','.').astype(float)
onboard_df['FuelTemperature'] = onboard_df['FuelTemperature'].str.replace(',','.').astype(float)
onboard_df['Throttle'] = onboard_df['Throttle'].str.replace(',','.').astype(float)
```
